Checking the count of food items for each category

```python
plt.figure(figsize=[10,6])
sns.countplot(y='Category',data=df_mcd)
plt.xticks(rotation=20);
plt.tight_layout()
```

Lets look at the calorie distribution across categories using a boxplot. This will help understand which categories have more calorically dense items

```python
plt.figure(figsize=[10,6])
sns.boxplot(y='Category',x='Calories',data=df_mcd)
plt.xticks(rotation=20);
plt.tight_layout()
```

The smoothies & shakes category is the clear winner in terms of number of calories. Breakfast, beef&pork and chicken&fish have a similar distribution and come second.

Further, we can investigate which categories contribute to a higher daily percentage of various unhealthy components like fat, sodium, sugar and cholesterol among others.

To start with, we can look at the distribution of daily fat contributions across this category.

```python
plt.figure(figsize=[10,6])
sns.boxplot(y='Category',x='Total Fat (% Daily Value)',data=df_mcd)
plt.xticks(rotation=20);
plt.tight_layout()
```

Breakfast, beef&pork and chicken&fish appears to have items that contribute to a higher daily permitted total fat percentage.

Next, lets make a similar comparison for cholesterol as a percentage of daily caloric intake

```python
plt.figure(figsize=[10,6])
sns.boxplot(y='Category',x='Cholesterol (% Daily Value)',data=df_mcd)
plt.xticks(rotation=20);
plt.tight_layout()
```

The breakfast section is clearly the category with many items contributing to a larger percentage of daily allowed cholesterol.

Next, up lets consider distribution of sugar among categories

```python
plt.figure(figsize=[10,6])
sns.boxplot(y='Category',x='Sugars',data=df_mcd)
plt.xticks(rotation=20);
plt.tight_layout()
```

Unsurprisingly, smoothies and shakes have the highest contribution, followed by coffee/tea, beverages as well as desserts.
Next, consider contributions to daily sodium.

```python
plt.figure(figsize=[10,6])
sns.boxplot(y='Category',x='Sodium (% Daily Value)',data=df_mcd)
plt.xticks(rotation=20);
plt.tight_layout()
```

Breakfast items, beef&pork as well as chicken&fish have a higher contribution to daily sodium percentages. 
Finally, consider carbohydrate percentages

```python
plt.figure(figsize=[10,6])
sns.boxplot(y='Category',x='Carbohydrates (% Daily Value)',data=df_mcd)
plt.xticks(rotation=20);
plt.tight_layout()
```

Summary:
 1. Smoothies & shakes have an overall higher calorie content followed by breakfast, beef&pork and chicken&fish categories
 2. Breakfast, beef&pork as well as chicken&fish categories have a higher median value for % contributions to daily calorie intake
 3. Among daily cholesterol percentages, breakfast items have a large number of items beyond a 25% daily contribution value
 4. Comparing sugars, smoothies&shakes, coffe&tea, beverages and desserts have a higher median sugar values among categories
 5. Looking at daily sodium contributions, breakfast, beef&pork as well as chicken&fish have higher median values 
 6. Lastly, among carbohydrate percentages, smoothies and shakes have the highest median value among categories

```python
#Define dataframes for each category
cat_dict = {}
category = df_mcd['Category'].value_counts().index.to_list()
for a in category:
    cat_dict[a] = df_mcd[df_mcd['Category']==a]
```

```python
df_mcd['Category'].value_counts()
```

Next, we can find items with maximum values in terms of carbohydrates, fats, cholesterols and so on among the 9 categories

```python
max_comp = df_mcd.groupby(['Category'],as_index=False).max()[['Category','Calories','Carbohydrates (% Daily Value)',
                                                   'Total Fat (% Daily Value)','Cholesterol (% Daily Value)',
                                                   'Sodium (% Daily Value)','Sugars']]
max_comp.columns = ['Category','Calories','max_pcnt_carb','max_pcnt_tot_fat','max_pcnt_chstrl',
                    'max_pcnt_sdm','max_sgr']
max_comp
```

```python
df_mrg = df_mcd.merge(max_comp,left_on='Category',right_on='Category')
```

```python
df_mcd.columns
```

```python
df_tot_fat = df_mrg.loc[(df_mrg['max_pcnt_tot_fat']==df_mrg['Total Fat (% Daily Value)']),
                        ['Category','Item','Total Fat (% Daily Value)']]
df_tot_fat.sort_values('Total Fat (% Daily Value)',ascending=False)
```

Chicken nuggets (although with a large serving size) exceed the daily fat allowance followed by the big breakfast meal

```python
df_tot_chstrl = df_mrg.loc[(df_mrg['max_pcnt_chstrl']==df_mrg['Cholesterol (% Daily Value)']),
                        ['Category','Item','Cholesterol (% Daily Value)']]
df_tot_chstrl.sort_values('Cholesterol (% Daily Value)',ascending=False)
```

Big breakfast meals account for the highest daily cholesterol percentages among all categories.

```python
df_sug = df_mrg.loc[(df_mrg['max_sgr']==df_mrg['Sugars']),
                        ['Category','Item','Sugars']]
df_sug.sort_values('Sugars',ascending=False)
```

Smoothies and shakes top the list for content of sugars. Most of the list is dominated by some kind of a beverage

```python
df_sod = df_mrg.loc[(df_mrg['max_pcnt_sdm']==df_mrg['Sodium (% Daily Value)']),
                        ['Category','Item','Sodium (% Daily Value)']]
df_sod.sort_values('Sodium (% Daily Value)',ascending=False)
```

Chicken nuggets again account for the highest sodium content followed by the big breakfast meal.
